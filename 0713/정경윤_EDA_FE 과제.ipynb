- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12

# 결측치 퍼센트를 데이터프레임으로 나타내기 
missing_percentage = df.isnull().sum() / len(df) 

# 결측치가 25% 이상인 경우 column drop 
df = df.drop(columns=missing_percentage[missing_percentage >= 0.25].index)

# 결측치가 10% 이상 25% 미만인 경우 row drop 
col_name = missing_percentage[(missing_percentage >= 0.10) & (missing_percentage < 0.25)].index.tolist()
for col in col_name:
    df = df.dropna(subset=[col], axis = 0)

# 결측치 10% 미만인 경우 중앙값 대체 
df = df.fillna(df.median())

# 결측치 확인하기 
df.info()
```

```python
#Q13-1) 가격에 대한 지표 변수 생성하기 
# 지역 평균 가격 대비 높은 가격인지 낮은 가격인지 알 수 있는 변수 필요 

# 가격의 범위 확인하기 
sns.boxplot(x="neighborhood", y="price", data = df)

# neighborhood로 group해서 price 평균
price = df.groupby(['neighborhood'])['price'].mean()
price
```

```python
# 지역 평균 가격 대비 높은지 낮은지 알 수 있는 변수 만들기 

regional_price = []

for index, row in df.iterrows():
    a = row['price']
    neighborhood = row['neighborhood']
    
    if neighborhood == 'Bronx':
        if a > price[0]:
            regional_price.append('higher')
        else:
            regional_price.append('lower')
    
    elif neighborhood == 'Brooklyn':
        if a > price[1]:
            regional_price.append('higher')
        else:
            regional_price.append('lower')
        
    elif neighborhood == 'Manhattan':
        if a > price[2]:
            regional_price.append('higher')
        else:
            regional_price.append('lower')
                   
    elif neighborhood == 'Queens':
        if a > price[3]:
            regional_price.append('higher')
        else:
            regional_price.append('lower')
                 
    elif neighborhood == 'Staten Island':
        if a > price[4]:
            regional_price.append('higher')
        else:
            regional_price.append('lower')

df['regional_price'] = regional_price
df.head()
```

```python
#Q13-2)
# 평균 이상의 리뷰 평점을 가진 경우, 우수하다는 지표 변수를 생성하여 직관적으로 평점의 정도를 파악할 수 있다 

mean_score = df['review_scores_rating'].mean()

good_review = []

for a in df['review_scores_rating']:
    if a > mean_score:
        good_review.append('good')
    else:
        good_review.append('bad')
        
df['good_review'] = good_review
df.head()
```

```python
#Q13-3)
# 레이블이 문자열로 되어 있으면 분석에 그대로 사용하기 어렵다. 숫자로 바꿔주는 레이블 인코딩 과정을 거쳐야 하기에 명목형 변수 중 하나인 room_type을 레이블인코딩했다. 

# 라벨 인코더 인스턴스 변수 만들기  
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# encoder에 room_type 변수 fitting 
encoded = encoder.fit_transform(df['room_type'])
df['room_type'] = encoded

df.room_type.value_counts()
# 0 : Entire home/apt
# 1 : Hotel room 
# 2 : Private room 
# 3 : Shared room 
```
